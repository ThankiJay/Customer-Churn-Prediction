## 1. Adding Duplicate entries

```python
# Randomly selecting rows and duplicating them
duplicate_rows = data.sample(n = 100, replace = True)
```

```python
data = pd.concat([data,duplicate_rows], axis=0)
```

```python
data.duplicated().value_counts()
```

# 2. Adding outliers to the data

```python
# Assuming you already have your DataFrame 'df'

# Set a random seed for reproducibility (optional)
np.random.seed(0)

# Randomly select 15 row indices to set 'MonthlyCharges'
num_rows = data.shape[0]
num_changes = 17

for _ in range(num_changes):
    row_idx = np.random.randint(0, num_rows)  # Random row index
    new_monthly_charge = np.random.uniform(300.20, 435.85)  # Generate a random charge
    data.at[row_idx, 'MonthlyCharges'] = new_monthly_charge

# Check the DataFrame with updated 'MonthlyCharges'
data.describe()
```

# Creating inconsitencies in data

```python
# randomly set 7 rows with column 'TotalCharges' values float from to string 
import pandas as pd
import numpy as np

# Assuming you already have your DataFrame 'df'

# Set a random seed for reproducibility (optional)
np.random.seed(0)

# Randomly select 7 row indices to set 'TotalCharges'
num_rows = data.shape[0]
num_changes = 7

for _ in range(num_changes):
    row_idx = np.random.randint(0, num_rows)  # Random row index
    new_total_charge = float(data.at[row_idx, 'TotalCharges'])  # Convert to float
    data.at[row_idx, 'TotalCharges'] = str(new_total_charge)  # Convert to string

# Check the DataFrame with updated 'TotalCharges'
data.head()
```

```python
data.info()
```

# Adding Missing Data

```python
import pandas as pd
import numpy as np

# Assuming you already have your DataFrame 'df'

# Set a random seed for reproducibility (optional)
np.random.seed(0)

# Randomly select 17 row indices to set 'OnlineSecurity' to NaN
num_rows = data.shape[0]
num_changes = 17

for _ in range(num_changes):
    row_idx = np.random.randint(0, num_rows)  # Random row index
    data.at[row_idx, 'OnlineSecurity'] = np.nan  # Set the value to NaN

# Check the DataFrame with 'OnlineSecurity' values set to NaN
data.head()
```

# Updating Yes/No Columns to 1/0

```python
columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']
mapping = {'Yes': 1, 'No': 0}

for column in columns:
    data[column] = data[column].map(mapping)
```
